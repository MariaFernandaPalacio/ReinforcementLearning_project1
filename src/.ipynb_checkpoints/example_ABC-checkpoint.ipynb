# Ejemplo de un agente SARSA resolviendo el entorno ABC

```python
#!pip install moviepy
```

```python
from utils.performer import Performer
from utils.interpreters import id_state
```

En primer lugar, creamos los parámetros del agente y luego creamos el objeto `Performer`. Esta clase contiene los métodos que vamos a utilizar para que los agentes resuelvan los entornos.

```python
agent_parameters = {'nA':2,
                    'nS':3,
                    'gamma':0.8,
                    'epsilon':0.1,
                    'alpha':0.1}
perf = Performer(env_name='ABC',
                 env_parameters=None,
                 state_interpreter=id_state,
                 agent_name='SARSA',
                 agent_parameters=agent_parameters)
```

Para verificar que el agente y el entorno están interactuando correctamente, podemos usar el método `run()`, el cual mostrará al agente actuando sobre el entorno. Dado que el render del entorno ABC es muy poco informativo, pues solo muestra el estado en el que está el agente, vamos a correr en modo para cazar errores usando el parámetro `visual=False`. Usualmente corremos este método sin argumentos, pues por defecto el valor es `True`:

```python
perf.run(visual=False)
```

Corremos el método de entrenamiento, `train()`, para decirle al agente que aprenda a resolver el entorno. Observe que se obtiene un dataset con el progreso del agente y una gráfica de la recompensa total por episodio. Si el agente está aprendiendo, la recompensa debería aumentar progresivamente. El agente (política y valores Q) se guarda en un archivo json al final del entrenamiento.

```python
perf.train()
```

Si queremos ver el desempeño del agente "en producción", es decir, sin que este pueda deambular explorando acciones aleatorias, podemos usar el método `test()`. Obtendremos el promedio de la recompensa total sobre 100 episodios y el correspondiente histograma:

```python
perf.test()
```

Finalmente, podemos usar el método `sweep()` para correr un barrido de parámetro. En este caso hemos escogido el parámetro de exploración $\epsilon$, el cual tomará valores 0, 0.01 y 0.1. Como resultado obtenemos una gráfica con la recompensa promedio por episodio, para cada uno de los valores del barrido:

```python
perf.sweep(parameter='epsilon',
           values=[0, 0.01, 0.1],
           num_simulations=100)
```

```python
agent_parameters = {'nA':2,
                    'nS':3,
                    'gamma':0.8,
                    'epsilon':0.01,
                    'alpha':0.1}
perf = Performer(env_name='ABC',
                 env_parameters=None,
                 state_interpreter=id_state,
                 agent_name='Q_learning',
                 agent_parameters=agent_parameters)
```

```python
perf.train()
```

- La imagen arrojada por el método test().

```python
perf.test()
```

- Comparación entre Q-learning y SARSA

```python
perf.sweep(parameter='epsilon',
           values=[0, 0.01, 0.1],
           num_simulations=100)
```

En las anteriores graficas vemos el progreso del aprendizaje que tiene Q_learning y SARSA en el entorno ABC, claramente el mejor desempeño en ambos casos son con epsilon de 0 seguido de un epsilon de 0.01.

- Use el método compare test() de la clase Performer para obtener un histograma de la recompensa total por episodio por 100 episodios.

```python
agent_vs_parameters = {
    'nA': 2,           # Número de acciones
    'nS': 3,           # Número de estados
    'gamma': 0.8,      # Factor de descuento
    'epsilon': 0.01,   # Valor de epsilon para epsilon-greedy
    'alpha': 0.1       # Tasa de aprendizaje
}
perf.compare_test(agent_vs_name='SARSA', 
                  agent_vs_parameters=agent_vs_parameters, 
                  num_episodes=100)
```

Al realizar la comparación entre Q_learning y SARSA vemos que en cuanto la suma promedio de las recompensas tuvo un mejor comportamiento el de SARSA teniendo una mayor frecuencia en la suma de las recompensas como se puede apreciar en la anterior grafica, pero en terminos generales tienen un desempeño similar ambos, con un promedio Q_learning de 8.77 y SARSA con 8.81.
