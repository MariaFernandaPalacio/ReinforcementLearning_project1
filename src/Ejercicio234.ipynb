## Ejercicio 2

Resolver Triqui con Q-learning. El entorno se puede cargar con la claseTriquiEnv del módulo environments.envs. Aquí es esencial usar el interprete de estado triqui interpreter del módulo utils.interpreters. Su respuesta debe incluir:

```python
from utils.performer import Performer
from utils.interpreters import id_state,triqui_interpreter, gym_interpreter1
```

- Una descripción del entorno:

En el entorno de triqui es un juego altamente conocido, donde vamos a contar con un tablero cuadrado 3x3, es decir consta de 9 casillas, dispuestas en una cuadrícula de 3 filas y 3 columnas. 

El entorno involucra dos jugadores donde uno utiliza la ficha "X" y el otro que utiliza la ficha "O". Cada jugador realiza movimientos alternativos. Así que el entorno va a contar con un total de 9 acciones.

Por otro lado,  en el triqui hay 9 celdas, y cada celda puede estar vacía, ocupada por X o ocupada por O. Esto da un total de 3^9 posibles configuraciones, lo cual va a ser el número de estados.

- Los parámetros de entrenamiento:

    Número de acciones = 9
    
    Número de estados = $3^9$
    
    Gamma = 0.8
    
    Epsilón = 0.1
    
    Alpha = 0.1

```python
agent_parameters = {'nA':9,
                    'nS':19683,
                    'gamma':0.8,
                    'epsilon':0.1,
                    'alpha':0.1}
perf = Performer(env_name='TriquiEnv',
                 env_parameters=None,
                 state_interpreter=triqui_interpreter,
                 agent_name='Q_learning',
                 agent_parameters=agent_parameters)
```

```python
perf.run(visual=True)
```

```python
perf.train()
```

```python
perf.test()
```

```python
agent_vs_parameters = {'nA':9,
                    'nS':19683,
                    'gamma':0.8,
                    'epsilon':0.1,
                    'alpha':0.1}
perf.compare_test(agent_vs_name='SARSA', 
                  agent_vs_parameters=agent_vs_parameters, 
                  num_episodes=100)
```
